Columns in the dataset

Removing unwanted columns

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

Extracting some information about the data

```python
df.info()
```

Removing duplicates

```python
df.drop_duplicates(inplace = True)
df.shape
```

Cleaning the rate column 

```python
df['rate'].unique()
```

Removing 'NEW', '-', and '/5' from the rate column

```python
def handlerate(value):
  if (value == 'NEW' or value == '-'):
    return np.nan
  else:
    value = str(value).split('/')
    value = value[0]
    return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

Finding null values

```python
df.rate.isnull().sum()
```

Filling null values with mean in the rate column

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

Dropping null values

```python
df.dropna(inplace = True)
df.head()
```

```python
df.info()
```

Renaming some lengthy column names

```python
df.rename(columns = {'approx_cost(for two people)': 'Cost for two', 'listed_in(type)': 'Type', 'listed_in(city)': 'City'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

Since here Location and City are similar columns, let's drop city column

```python
df = df.drop(['City'], axis = 1)
```

```python
df.head()
```

```python
df['Cost for two'].unique()
```

Since some values of Cost for two column are having a comma, so it's better to remove them

```python
def handlecomma(value):
  value = str(value)
  if ',' in value:
    value = value.replace(',', '')
    return float(value)
  else:
    return float(value)

df['Cost for two'] = df['Cost for two'].apply(handlecomma)
df['Cost for two'].unique()
```

```python
df.head()
```

Cleaning the rest_type column

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
```

```python
rest_types_less_than_1000 = rest_types[rest_types<1000]
rest_types_less_than_1000
```

Making rest_types_less_than_1000 in frequency as others

```python
def handle_rest_type(value):
  if (value in rest_types_less_than_1000):
    return 'Others'
  else:
    return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
df['location'].value_counts()
```

Similarly, here we need to clean the location column as well

```python
location = df['location'].value_counts(ascending = False)
```

```python
location_less_than_300 = location[location<300]
```

```python
def handle_location(value):
  if(value in location_less_than_300):
    return 'others'
  else:
    return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
df['cuisines'].value_counts()
```

Similarly, we need to clean the cuisines column as well

```python
cuisines = df['cuisines'].value_counts(ascending = False)
cuisines_less_than_100 = cuisines[cuisines<100]

def handle_cuisines(value):
  if (value in cuisines_less_than_100):
    return 'others'
  else:
    return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df['votes'].value_counts()
```
